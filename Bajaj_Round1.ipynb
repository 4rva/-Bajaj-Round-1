**Q.4** Share percentage of missing values, for the columns firstName, lastName and DOB from patientDetails.

Note: Missing values are defined as blank and empty strings. Percentage rounded off to 2 decimal.

```python
fn = (df['patientDetails.firstName'].isnull().sum())/31*100
fn = round(fn, 2)
ln = (df['patientDetails.lastName'].isnull().sum())/31*100
ln = round(ln, 2)
bd = (df['patientDetails.birthDate'].isnull().sum())/31*100
bd = round(bd, 2)
```

```python
print("The Answer to Q4 is", fn, ",", ln, ",", bd)
```

**Q.9**Find the percentage distribution of total active and inactive medicines.

```python

all_medications = [med for sublist in df['consultationData.medicines'] for med in sublist if isinstance(sublist, list)]
active_count = sum(1 for med in all_medications if med.get('isActive') == True)
inactive_count = sum(1 for med in all_medications if med.get('isActive') == False)

total = active_count + inactive_count
active_percentage = round((active_count / total) * 100, 2)
inactive_percentage = round((inactive_count / total) * 100, 2)


print(active_percentage, ",", inactive_percentage,)
```
